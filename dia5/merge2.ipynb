```python
import pandas as pd
```

```python
df_customer = pd.read_csv('customers.csv', sep=";")
df_customer
```

```python
df_transactions = pd.read_excel('transactions.xlsx')
df_transactions
```

```python
df_transactions_product = pd.read_parquet('transactions_cart.parquet')
df_transactions_product
```

```python
df_join = (df_transactions.merge(df_customer,
                                how="inner",
                                left_on="IdCustomer",
                                right_on="UUID",
                                suffixes=["_transacao", "_cliente"])
                          .merge(df_transactions_product,
                                 how='inner',
                                 left_on="UUID_transacao",
                                 right_on="IdTransaction")
                                 )

df_join
```
