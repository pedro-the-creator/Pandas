```python
import pandas as pd
```

```python
df = pd.read_excel('transactions.xlsx')
df
```

```python
df_summary = df.groupby(["IdCustomer"])["Points"].sum()
df_summary.reset_index()
```

```python
(df.groupby(["IdCustomer"])
   .agg({ "Points": 'sum',
          "UUID": "count",
          "DtTransaction": "max"})
    .rename(columns={
            "Points":"Valor",
            "UUID": "Frequencia",
            "DtTransaction":"UltimaData"
            })
   .reset_index())
```

```python
import datetime

data1 = df["DtTransaction"][0]
now = datetime.datetime.now()

(now - data1).days
```

```python
(datetime.datetime.now() - df["DtTransaction"][0]).days
```

```python
def recencia(x):
    diff = datetime.datetime.now() - x.max()
    return diff.days


(df.groupby(["IdCustomer"])
   .agg({ "Points": 'sum',
          "UUID": "count",
          "DtTransaction": ['max', recencia]
          })
    .rename(columns={
            "Points":"Valor",
            "UUID": "Frequencia",
            "DtTransaction":"UltimaData"
            })
   .reset_index())
```
